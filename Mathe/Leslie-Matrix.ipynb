## Leslie-Matrix

Im folgenden betrachten wir die Leslie-Matrix

$$L = \begin{pmatrix}
0.1 & 1.4 & 0.5 \\
0.5 & 0 & 0 \\
0 & 0.8 & 0
\end{pmatrix}$$

Den Populationszustand $x_{t+1}$ können wir aus $x_t$ wie folgt berechnen:
$$ x_{t+1} = L \cdot x_t
$$

Wir möchten nun eine Funktion in `python` implementieren, welche aus einer Matrix $L$ und einem Populationszustand $x_0$ die nächsten $n$ Populationszustände berechnet und zurück gibt:

```python
def pop_entw(matrix, x_start, n):
    # Erstelle eine Liste 'pop' welche den ersten Populationszustand erhält. Diese erweitern wir nun n-mal um den jeweils nächsten
    # Populationszustand
    pop = [x_start]

    # x_n beschreibt immer den aktuellen Populationszustand. Daher initialisieren wir 'x_n' mit 'x_start'
    x_n = x_start

    # Die range()-Fkt. wird hier verwendet, um die Rechnung n-mal auszuführen. 
    for i in range(1,n):
        x_n = matrix@x_n # Neuen Populationszustand berechnen
        pop.append(x_n) # 'pop'-Liste um nächsten Pop.-zustand erweitern.

    # Liste mit den Pop.-zuständen zurück geben.
    return pop
```

Die erstellte Funktion wenden wir nun auf die Leslie-Matrix von oben an. Als Startzustand wählen wir $x_0 = (19,10,10)^T$.

```python
import numpy as np

L = np.array([[.1, 1.4, .5], [.5, 0, 0], [0, .8, 0]])
x_0 = np.array([19,10,10])

print(pop_entw(L, x_0, 20))
```

Diese Pop.-zustände können wir auch in einem Plot darstellen:

```python
import matplotlib.pyplot as plt

x_n = pop_entw(L, x_0, 15)
n = list(range(0,15))

plt.scatter(n, [x[0] for x in x_n], label='AK 1')
plt.scatter(n, [x[1] for x in x_n], label='AK 2')
plt.scatter(n, [x[2] for x in x_n], label='AK 3')
plt.legend()
plt.show()
```
