```python
def pop_entw(matrix, x_start, n):
    pop = [x_start]
    x_n = x_start
    for i in range(1,n):
        x_n = matrix@x_n
        pop.append(x_n)
    return pop
```

```python
import numpy as np

M = np.array([[.1, 1.4, .5], [.5, 0, 0], [0, .8, 0]])
x_0 = np.array([19,10,10])

print(pop_entw(M, x_0, 20))
```

```python
import matplotlib.pyplot as plt

x_n = pop_entw(M, x_0, 15)
n = list(range(0,15))

plt.plot(n, [x[0] for x in x_n], label='AK 1')
plt.plot(n, [x[1] for x in x_n], label='AK 2')
plt.plot(n, [x[2] for x in x_n], label='AK 3')
plt.legend()
plt.show()
```
