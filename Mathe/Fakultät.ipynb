## Fakultät

### 1. Rekursive Funktion

```python
def fak_rek(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n*fak_rek(n-1)
    
fak_rek(7)
```

### 2. Funktion mit for-Schleife

```python
def fak(n):
    n_fak = n
    for i in range(1,n):
        n_fak=n_fak*i
    return n_fak

fak(7)
```

### 3. Fkt. mit while-Schleife

```python
def fak2(n):
    i = 1
    n_fak = 1
    while i <= n:
        n_fak = i*n_fak
        i = i+1
    return n_fak

fak2(7)
```
