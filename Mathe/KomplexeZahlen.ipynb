## Der Datentyp complex für komplexe Zahlen

Python stellt auch einen eingebauten Datentyp für komplexe Zahlen zur Verfügung, d.h. wir können die Menge $\mathbb C$ ohne besondere Pakete abbilden: Dieser Datentyp heißt <code>complex</code>. Wichtig ist: Die imaginäre Einheit $i$ wird in Python mit <code>j</code> bezeichnet.

Wir können die Zahl $z = a + i*b$ in Python durch <code>z = a + bj</code> anlegen, man beachte im folgenden Beispiel, wie genau der imaginäre Anteil geschrieben werden muss.

Für den Datentyp <code>complex</code> sind die gängigen Grundrechenarten bereits definiert, ebenso spezielle Operationen auf komplexen Zahlen wie die komplexe Konjugation, d.h. das Umdrehen des Vorzeichens des imaginären Anteils:

```python
z1 = 2 + 3j
z2 = 1 + 1j

print(z1+z2)
print(z1*z2)
print(z1.conjugate()) 

# Berechnung des Betragsquadrats:
print("|z1|^2 =", z1*z1.conjugate())

# Die imaginäre Einheit quadriert ergibt -1:
print("j^2 =", 1j*1j)

# Auf Real- bzw. Imaginärteil zugreifen:
print('Realteil von z1:', z1.real)
print('Imaginärteil von z1:', z1.imag)
```

Mit denselben Funktionen können auch komplexe Eigenwerte und Eigenvektoren berechnet werden:

```python
import numpy as np

M = np.array(([4,-3],[3,4]))

eigenwerte, eigenvektoren = np.linalg.eig(M)
print(eigenwerte)
print(eigenvektoren)
```

### Visuelle Darstellung für die Rechenregeln komplexer Zahlen

```python
import matplotlib.pyplot as plt


plt.quiver(z1.real, z1.imag, angles='xy', scale_units='xy', scale=1, label='z1', color='darkblue')
plt.quiver(z2.real, z2.imag, label='z2', angles='xy', scale_units='xy', scale=1, color='lightblue')
plt.quiver((z1*z2).real, (z1*z2).imag, label='z1*z2', angles='xy', scale_units='xy', scale=1, color='forestgreen')
plt.quiver(z1.conjugate().real, (z1.conjugate()).imag, label='z1 konj.', angles='xy', scale_units='xy', scale=1, color='gold')
plt.axhline(color='grey') # x-Koordinatenachse
plt.axvline(color='grey') # y-Koordinatenachse
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.legend()
plt.show()
```
