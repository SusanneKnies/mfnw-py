## Der Datentyp bool für logische Variablen
Ein logischer Datentyp kennt nur zwei Zustände: **wahr** und **falsch**. In der Informatik nennt man dies **binär**, somit ist dieser Datentyp sehr speichereffizient, weil nur eine binäre Null (mit der semantischen Bedeutung "falsch") oder eine binäre Eins (mit der semantischen Bedeutung ("wahr"), also nur ein sogenanntes **Bit** abgespeichert werden muss. 

Die Festlegung, ob Null=falsch oder andersherum, ist offenbar willkürlich.

Zur Einordnung: Ein Bit ist die kleinste Speichereinheit im Computer. 8 Bit formen ein Byte. Üblicherweise sind wir in weit größeren Bereichen unterwegs, bspw. beim monatlichen Datenvolumen unserer Mobilfunkverträge oder beim Speicherplatz unserer Telefone im Bereich von GigaByte. Da 1 GB 1.000.000.000 (eine Milliarde) Byte sind, ist klar, dass ein bit in der Tat sehr wenig Speicher belegt.

In Python wird ein logischer Datentyp durch die speziellen Zustände <code>True</code> (wahr / richtig) und <code>False</code> (unwahr / falsch) gekennzeichnet, die Großschreibung ist hier wichtig. Damit sind diese beiden Bezeichner keine zulässigen Namen für Variablen, und wir müssen uns nicht merken, was davon eine Null und was eine Eins ist.

Die Variablen haben den Datentyp <code>bool</code>, benannt nach dem englischen Mathematiker, Logiker und Philosoph George Boole (1815-1864). Wir sprechen auch von **Boolschen Variablen**.


```python
variable1 = True
variable2 = False
print("variable1 = {}, Typ = {}".format(variable1, type(variable1)))
print("variable2 = {}, Typ = {}".format(variable2, type(variable2)))
```

### Boolsche Variablen kennen wir implizit schon: <code>while</code> Schleifen

Diese Schleifen, präziser ihre Abbruchbedingung, basiert auf dem Datentyp <code>bool</code>:

```python
i = 5
while i >= 1:
    print(i)
    i = i - 1
```

Wir sehen hier eine **logische Bedingung** der Form <code>i >= 1</code>. Um dies auch von Python bestätigen zu lassen, schauen wir einfach einmal:

```python
i = 5
print(i>=1)
print(type(i>=1))
```

Wir können das Beispiel auch wie folgt schreiben, um dies expliziter hervorzuheben:

```python
i = 5
condition = True
print(i)
print(condition)
print("-"*5)
while condition:
    i = i - 1
    # Klammerung als didaktische  Hervorhebung
    condition = (i >= 1)
    print(i)
    print(condition)
    print("-"*5)
```

### Operationen mit <code>bool</code>-Variablen
Mit <code>bool</code>-Variablen können logische Verknüpfungen sehr einfach realisiert werden. Drei elementare logische Verknüpfungen, die aus der Mengenlehre und somit der Schulmathematik bekannt sind, werden in Python folgendermaßen repräsentiert:

```python
a = True
b = False
# NOT: genau dann wahr, wenn a nicht wahr ist (False)
print("  not a =", not a)
# AND: genau dann wahr, wenn beide Seite wahr sind
print("a and b =", a and b)
# OR: genau dann wahr, wenn mindestens eine der beiden Seiten wahr ist
print(" a or b =", a or b)
```

### Wahrheitstabellen für die logischen Verknüpfungen <code>and</code> und <code>or</code>
Der folgende Codeschnipsel verdeutlicht neben den Wahrheitstabellen, wie die formatierte Ausgabe für den <code>bool</code>-Datentyp funktioniert:

```python
print("  a   |    b  || a and b | a or b")
print("------+-------++---------+-------")
print("False | False ||  {}  |  {}".format(False and False, False or False))
print("False | True  ||  {}  |  {}".format(False and True, False or True))
print("True  | False ||  {}  |  {}".format(True and False, True or False))
print("True  | True  ||  {}   |  {}".format(True or True, True or True))
```

Die Reihenfolge der Operationen ist wichtig. Es gilt gemäß der Mengenlehre: "<code>not</code> geht vor <code>and</code> und <code>or</code>". Zudem können die Ausdrücke geklammert werden. Es ist empfehlenswert, sich zunächst die Ausgabe gemäß dieser Rechenregeln und der obigen Wahrheitstabelle zu überlegen, bevor das Beispiel ausgeführt wird.

```python
a = True
b = False
print("a =", a)
print("b =", b)
print("not a and b   =", not a and b)
print("not (a and b) =", not (a and b))
print("not a or b    =", not a or b)  # not geht natürlich auch vor "or"
print("not (a or b)  =", not (a or b))
```

Um uns endgültig davon zu überzeugen, dass Python die klassische Logik vernünftig wiederspiegelt, programmieren wir eines der sogenannten [de-Morganschen Gesetze](https://de.wikipedia.org/wiki/De-morgansche_Gesetze) nach, das im Prinzip sagt, wie geklammerte logische Ausdrücke "ausmultipliziert" werden:
$$ \neg(a \wedge b) = \neg a \vee \neg b $$
Hierbei ist $\neg$ die mathematische Notation für die Negation (<code>not</code>), $\wedge$ ist das Symbol für <code>and</code>, und $\vee$ das Symbol für <code>or</code>. Man sollte sich klarmachen, warum auf der rechten Seite dieser Gleichung keine Klammerung nötig ist.

```python
a = True
b = False
print(not (a and b), '=', not a or not b)
# Impro-Quiz: Warum keine Klammerung rechts?
```

### Vorausschauende Auswertung
Die logischen Operatoren <code>and</code> und <code>or</code> sind in Python "intelligent" / "vorausschauend":
* <code>a and b</code>: Wenn <code>a</code> falsch ist, wird <code>b</code> nicht evaluiert.
* <code>a or b</code>: Wenn <code>a</code> wahr ist, wird <code>b</code> nicht evaluiert.

Dies ist bei komplexere Ausdrücken (siehe spätere Lerneinheiten) durchaus sinnvoll, wenn die Auswertung eines logischen Ausdrucks aufwendiger ist als in unseren aktuellen Spielzeugbeispielen. Es kann effizient sein, "einfache" Bedingungen zuerst zu überprüfen. 


### Fortgeschrittenes Thema: unintuitives Verhalten bei schlechter Programmierung

Die vorausschauende Auswertung logischer Ausdrücke kann gefährlich sein, wenn man dieses Verhalten nicht kennt. Wir verdeutlichen das wie üblich mit Beispielen, bevor wir eine Regel ableiten:

```python
some_variable = True
F = some_variable or i_am_not_defined
print(F)    # die nicht definierten Variable wird durch das Vorausschauen nicht erkannt:
some_variable = False

# dasselbe nochmal mit AND
F = some_variable and i_am_not_defined
print(F)
```

Diese Eigenart von Python kann zu sehr schwer identifizierbaren Fehlern führen, insbesondere dann, wenn sehr komplexe Operationen notwendig sind um <code>some_variable</code> zu bestimmen.

Wenn allerdings der komplette Ausdruck ausgewertet wird, weil die vorausschauende Auswertung nicht greift, dann ist Python nett genug uns eine Fehlermeldung zu liefern:

```python
# wie eben, das zweite Argument von "or" wird ignoriert:
some_variable = True
F = some_variable or i_am_not_defined
print(F)

# Forcierung der Auswertung des zweiten Arguments:
some_variable = False
F = some_variable or i_am_not_defined
print(F)
```

```python
# so wird das repariert:

some_variable = False
i_am_now_defined = 'Hallo, ich bin definiert!'

F = some_variable or i_am_now_defined
print(F)
```

Das Beispiel verwirrt allerdings auch, denn die logische <code>or</code> Verknüpfung einer <code>bool</code>-Variable mit einer Zeichenkette ist die Zeichenkette. Bei <code>and</code> ist das andersherum:

```python
# so wird das repariert:

some_variable = False
i_am_now_defined = 'Hallo, ich bin definiert!'

F = some_variable and i_am_now_defined
print(F)
```

Als Auflösung nennen wir die folgende **Faustregel:** Alle Objekte in Python, die nicht <code>False</code>, 0 oder leer sind, ergeben bei einer Konvertierung zu <code>bool</code> den Wert <code>True</code>:

```python
s = 'some string'
print("bool('some string') =", bool(s))
s = ''
print("bool('') =", bool(s))
n = 10.2
print("bool(10.2) =", bool(n))
n = 0
print("bool(0) =", bool(n))
```

Umgekehrt entsprechen <code>True</code> und <code>False</code> den Zahlen 1 und 0:

```python
print("True  =", int(True))
print("False =", int(False))
```

Das kann zu "verwirrenden" Codezeilen und Programmabläufen führen, siehe auch das Beispiel weiter oben mit Zeichenketten:

```python
a = 5
print(a and True)  # a wird wie True behandelt
b = 0
print(True and b)  # Hier ist b False
print(bool(b))
```

**Merke:** Auch wenn eine Programmiersprache solche unintuitiven "Schweinereien" erlaubt, sollten wir uns sie gar nicht erst angewöhnen, und logische Ausdrücke immer explizit als <code>bool</code> typisieren.

### Vergleichsoperationen mit <code>int</code> Variablen
Vergleichsoperationen mit <code>int</code>-Variablen funktionieren genau so, wie wir es aus der Schulmathematik erwarten:
* echt kleiner: <code> \< </code>
* echt größer: <code> \> </code>
* kleiner gleich: <code> \<= </code>
* größer gleich: <code> \>= </code>
* gleich: <code> == </code>
* ungleich: <code> != </code>

Das Ergebnis einer Vergleichsoperation ist eine <code>bool</code> Variable:
    

```python
a, b = 5, 6
result = a < b
print("Datentyp: {}, Ergebnis: {}".format(type(result),result))
result = a != b
print(result)
print(a >= b)  # hier ohne den Umweg, das Ergebnis des Vergleichs in einer Variable zu speichern
```
