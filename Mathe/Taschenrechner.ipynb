### Kurzschreibweise für die elementaren Operationen
Anweisungen wie <code>a = a + 1</code> können in Python wie folgt abgekürzt werden:

```python
a = 5
a += 5    # Entspricht a = a + 5
a -= 5    # Entspricht a = a - 5
a *= 5    # Entspricht a = a * 5
a **= 5   # Entspricht a = a ** 5
a /= 5    # Entspricht a = a / 5
a //= 5   # Entspricht a = a // 5
```
