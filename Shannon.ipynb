# Shannon Index

```python
import csv

with open('Shannon_Index.csv') as csvfile:
    reader = csv.reader(csvfile)
    data = [row for row in reader]

colnames = data[0]
data = data[1:]
print(data)
```

```python
M1 = sum(int(row[1]) for row in data)
M2 = sum(int(row[2]) for row in data)
```

```python
p1 = [int(row[1])/M1 for row in data]
p2 = [int(row[2])/M2 for row in data]
```

```python
# test
print(sum(p1))
print(sum(p2))
```

```python
# Shannon Index berechnen
import math

H1 = -sum(p*math.log2(p) for p in p1 if p>0)
print(H1)

H2 = -sum(p*math.log2(p) for p in p2 if p>0)
print(H2)
```

```python
# Index wenn alle Arten gleich oft vorkommen
n3 = [30]*len(p1)
M3 = sum(n3)
p3 = [n/M3 for n in n3]

H3 = -sum(p*math.log2(p) for p in p3 if p>0)
print(H3)
```
