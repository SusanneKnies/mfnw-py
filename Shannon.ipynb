# Shannon Index (vgl. Blatt 7, Aufgabe 1, WS)

Im folgenden wollen wir die csv-Datei 'Shannon_Index' einlesen. Diese enthält eine Tabelle mit den Anzahlen der Individuen $n_i$ aller Makroinvertebraten aufgelistet, die in zwei Abschnitten P1 und P2 der Plane, einem Bach im Hohen Fläming (Brandenburg), gefunden wurden. Wir wollen nun für die zwei Abschnitte den Shannon-Index berechnen. Zur Erinnerung: Der Shannon Index $H_S$ ist ein aus der Informationstheorie stammendes Maß für die Diversität einer Biozönose, das genau diese beiden Eigenschaften besitzt:

$$ H_S = -\sum\limits_{i=1}^N p_i \text{ log}_2 \,\, p_i, \quad \text{mit } p_i = \frac{n_i}{M} \quad \text{und } M = \sum\limits_{i=1}^N n_i.$$

$N$ ist die Anzahl der gefundenen Arten, $n_i$ sind deren Individuenzahlen, $M$ ist die Anzahl der insgesamt gefundenen Individuen. Die Wahrscheinlichkeit bzw. relative Häufigkeit, ein Individuum der Art $i$ anzutreffen ist $p_i$.

### csv-Datei einlesen

Um eine csv-Datei einzulesen importieren wir die Bibliothek `csv`. Mit dem folgenden Code können wir dann die Datei einlesen. Die Variable `data` ist dann eine Liste mit allen Zeilen der csv-Datei. Da die erste Zeile die Spaltennamen enthält erstellen wir die Variable `colnames`, welche dann die Spaltennamen enthält. Danach überschreiben wir `data` mit `data[1:]`. Nun enthält `data` alle Zeilen der csv-Datei außer der ersten.

```python
import csv

with open('Shannon_Index.csv') as csvfile:
    reader = csv.reader(csvfile)
    data = [row for row in reader]

colnames = data[0]
data = data[1:]
print(data)
```

### Anzahl der Individuen (M) berechnen

Die i-te Liste in der Liste `data` enthält (der Reihenfolge nach):
- Den Namen der i-ten Gattung
- Die Anzahl der Individuen der i-ten Gattung im Abschnitt P1
- Die Anzahl der Individuen der i-ten Gattung im Abschnitt P2

Um die Anzahlen der Individuen je nach Planenabschnitt zu berechnen kann folgender Code verwendet werden:

```python
M1 = sum(int(row[1]) for row in data)
M2 = sum(int(row[2]) for row in data)
```

**Erklärung zum Code**
- `for row in data`: Es wird Zeile für Zeile durchgegangen
- `row[1]`, bzw. `row[2]`: Die Anzahl der Individuen ($n_i$) im ersten, bzw. zweiten Abschnitt der Plane
- `int()`: Beim Einlesen der csv-Datei wird jeder Eintrag als `string` eingelesen. Da wir jedoch die Summe aller $n_i$ berechnen wollen, müssen wir diese in einen `integer umwandeln`
- `sum()`: Berechnet die Summe aller $n_i$

### Relative Häufigkeit $p_i$ berechnen (als ÜA?)

Unten dann Code zur Überprüfung

```python
p1 = [int(row[1])/M1 for row in data]
p2 = [int(row[2])/M2 for row in data]
```

```python
# test
print(sum(p1))
print(sum(p2))
```

### Ebenfalls ÜA?

Man könnte auch die Rechnung für H1 komplett vorstellen und dann Berechnung für H2 als ÜA lassen (dann schon ab dem Schritt zur Berechnung von M2). Transfer wäre dann den richtigen row-index anzupassen

```python
# Shannon Index berechnen
import math

H1 = -sum(p*math.log2(p) for p in p1 if p>0)
print(H1)

H2 = -sum(p*math.log2(p) for p in p2 if p>0)
print(H2)
```
