### Möglichkeiten

```python
a = 'CT'
b = 'AC'

möglichkeiten = [aa+bb for aa in a for bb in b]
print(möglichkeiten)
```

### Allelen sortieren

```python
b_sorted = sorted(b)
print(b_sorted)

b_sorted = ''.join(b_sorted)
print(b_sorted)
```

```python
möglichkeiten = [''.join(sorted(allel)) for allel in möglichkeiten]
print(möglichkeiten)
```

### Doppelte Möglichkeiten aussortieren

```python
c = 'CT'

möglichkeiten_ac = [''.join(sorted(aa+cc)) for aa in a for cc in c]
print(möglichkeiten_ac)

möglichkeiten_ac = list(set(möglichkeiten_ac))
print(möglichkeiten_ac)

möglichkeiten_ac.sort()
print(möglichkeiten_ac)
```

```python
# Alternative

möglichkeiten_ac = [''.join(sorted(aa+cc)) for aa in a for cc in c]
print(möglichkeiten_ac)

new_list = []
for allel in möglichkeiten_ac:
    if allel not in new_list:
        new_list.append(allel)

print(new_list)
```

### Wahrscheinlichkeiten berechnen

```python
alle_möglichkeiten_ac = [''.join(sorted(aa+cc)) for aa in a for cc in c]

print(möglichkeiten_ac, alle_möglichkeiten_ac)

anzahl_allel = []
for allel in möglichkeiten_ac:
    p_i = alle_möglichkeiten_ac.count(allel)
    anzahl_allel.append(p_i)

print(möglichkeiten_ac, anzahl_allel)

wahrscheinlichkeiten = [p_i/sum(anzahl_allel) for p_i in anzahl_allel]
print(möglichkeiten_ac, wahrscheinlichkeiten)
```
