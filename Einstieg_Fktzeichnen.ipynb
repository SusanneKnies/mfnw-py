## Funktionen zeichnen mit Matplotlib
Im folgenden Abschnitt lernst du, wie man mathematische Funktionen mithilfe der Bibliotheken <code>numpy</code> und <code>matplotlib</code> visualisiert. Das Prinzip ist dabei immer gleich: Wir erstellen eine Liste von $x$-Werten und berechnen für jeden dieser Werte den entsprechenden $y$-Wert.

Im folgenden findest du <code>python</code>-Code um die Funktion $f(x)=x^2$ darzustellen:

Zuerst einmal laden wir dafür die Bibliotheken `numpy` und `matplotlib`. Diese stellen Funktionen zur Verfügung, welche nicht standardmäßig in `python` implementiert sind. Um sie zu laden verwendest du folgende Befehle:

```python
import matplotlib.pyplot as plt
import numpy as np
```

Achte darauf den ersten Code-Block auch auszuführen, um die Bibliotheken zu laden. Mit `as np` legen wir fest, dass die in numpy verfügbaren Funktionen mit `np.` abgerufen werden können. Nun können wir folgenden Code-Block ausführen:

```python
x = np.linspace(-1, 1, 100) # x-Achse in 100 Pkt. zwischen -1 und 1 unterteilen
```

Die in `numpy` verfügbare Funktion `linspace` erstellt gibt 100 geordnete Punkte zwischen -1 und 1 zurück, welche den gleichen Abstand zueiander haben.  Um nun die Funktion $f(x)=x^2$ kannst du folgenden Code verwenden:

```python
y = x**2 # zu jedem x-Wert zugehörigen y-Wert berechnen

plt.plot(x,y, color='blue', label='f(x)=x^2') # Pkt. plotten
plt.xlabel("x") # x-Achse beschriften
plt.ylabel("y")
plt.legend()
plt.show() # Erst durch diesen Befehl wird der Plot unten angezeigt
```

Hier findest du denselben Code. Versuche nun den Code so zu ändern, dass er folgende Bedingungen erfüllt:

- Der Graph zeigt die Funktion $f(x)=2x^3-x$.
- Die Funktion wird für x-Werte aus dem Intervall $[-2; 3]$ dargestellt.
- Die Funktionslinie ist rot dargestellt.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 100)
y = x**2

plt.plot(x,y, color='blue', label='f(x)=x^2')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# Lösung

x = np.linspace(-2, 3, 100)
y = 2*x**3-x

plt.plot(x,y, color='red', label='f(x)=2x^3-x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
