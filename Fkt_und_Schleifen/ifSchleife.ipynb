## Bedingte Ausführung mit if-elif-else
Um in einem Programm Entscheidungen treffen zu können, kann das <code>if</code>-Konstrukt verwendet werden, eine weitere **Kontrollstruktur** neben den <code>while</code> und <code>for</code> Schleifen. Wir erläutern den Syntax anhand einiger Beispiele, und erinnern daran, dass innerhalb von Kontrollstrukturen **Blöcke** durch Einrückungen realisiert und durch einen Doppelpunkt begonnen werden:

```python
number = int(input("Bitte eine Zahl eingeben: "))
if number <= 10:
    print("Die Zahl ist kleiner oder gleich 10")
```

```python
number = int(input("Bitte eine Zahl eingeben: "))
if number <= 10:
    print("Die Zahl ist kleiner oder gleich 10")
else:
    print("Die Zahl ist größer als 10")
```

```python
number = int(input("Bitte eine Zahl eingeben: "))
if number <= 10:
    print("Die Zahl ist kleiner oder gleich 10")
elif number % 2 == 0:
    print("Die Zahl ist größer als 10 und durch zwei teilbar")
else:
    print("Die Zahl ist größer als 10 und nicht durch zwei teilbar")
```

Bei der <code>if</code>-Abfrage wird stets der erste zutreffende Block ausgeführt. Es ist möglich, beliebig viele <code>elif</code>-Blöcke (kurz für else if) zu programmieren. Wichtig ist, dass nach dem <code>else</code>-Block die Kontrollstruktur beendet ist.

### Beispiel: Test auf Schaltjahr
Die folgenden Regeln legen fest ob ein Jahr $J \in \mathbb N$ ein Schaltjahr ist:
* Das Jahr muss durch 4 teilbar sein.
* Säkulärjahre, also die Jahre, die ein Jahrhundert abschließen sind keine Schaltjahre (Beispiel 1800, 1900, 2100 und 2200).
* Ist $j$ durch 400 teilbar, so ist es doch ein Schaltjahr.

1996 und 1992 sind beispielsweise Schaltjahre.

```python
year = int(input("Bitte ein Jahr eingeben: "))

if (year % 4 == 0 and
   (not year % 100 == 0 or year % 400 == 0)):
    print("Das Jahr {} ist ein Schaltjahr".format(year))
else:
    print("Das Jahr {} ist kein Schaltjahr".format(year))
```

Im Beispiel haben wir noch eine weitere neue Funktionalität eingeführt, die Code lesbarer machen kann, indem Anweisungen über mehrere Zeilen geschrieben werden. Hierbei müssen jedoch Klammern um den gesamten Ausdruck gesetzt werden. Alternativ kann eine neue Zeile auch mit einem Backslash angedeutet werden, wie der nächste Codeschnipsel zeigt.

Wir verabreden folgende Konvention: Wir versuchen die Zeilenlänge in unseren Programmen möglichst kurz zu halten. Dies dient vor allem der besseren Lesbarkeit des Codes. Als Faustregel sollte jede Zeile **nicht mehr als 80 Zeichen** enthalten!

```python
year = int(input("Bitte ein Jahr eingeben: "))

if year % 4 == 0 and \
   (not year % 100 == 0 or year % 400 == 0):
    print("Das Jahr {} ist ein Schaltjahr".format(year))
else:
    print("Das Jahr {} ist kein Schaltjahr".format(year))
```
