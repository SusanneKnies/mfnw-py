### Plotten funktionaler Zusammenhänge

Im Mathematik-Unterricht ist es oft ausgesprochen hilfreich, funktionale Zusammenhänge graphisch darzustellen. Dazu gibt es natürlich bereits 4598493847539 verschiedene Software-Pakete, so dass wir eigentlich nicht (neu) lernen müssen, wie dies in Python funktioniert. Das folgende Beispiel soll dennoch verdeutlichen, dass wir diese Aufgabe auch mit Python und einem passenden Paket erfüllen können, und zwar ohne großen Aufwand und innerhalb **einer** Umgebung.

Wenn Sie am Ende dieses Notebooks alle Beispiele gemeinsam im Blick haben, haben Sie eine erste Idee davon, wie groß die Bedeutung der letzten Aussage ist: Wir müssen keine Daten und Funktionalitäten aus unterschiedlichsten Lösungen mit hohem manuellen Aufwand zusammenführen, sondern können alles in einer einzigen Umgebung erreichen. Dies spart langfristig enorm Zeit, und ist den (einmaligen) Mehraufwand dieses Grundlagenkurses in Python mehr als wert.

Wie genau die verwendeten Python-Pakete **numpy** und **matplotlib** funktionieren, erklären wir in einer späteren Einheit dieses Kurses. Auch sollten Sie im Moment noch syntaktische Details ignorieren.



```python
import numpy as np
import matplotlib.pyplot as plt

# erzeugt äquidistante Auswertungspunkte im gegebenen Intervall zwischen -pi und pi
eval_points = np.linspace(-np.pi, np.pi, 300)

# wertet die Sinusfunktion in allen Auswertungspunkten gleichzeitig aus
values = np.sin(eval_points)

# plottet die generierten (x,y)-Paare (Paare aus Punkten auf der x-Achse und 
# zugehörigen Funktionsauswertungen  auf der y-Achse) als Polygonzug / Linienzug
plt.plot(eval_points,values)                   
      
# zeigt den Plot an
plt.show()
```
