### Plotten funktionaler Zusammenhänge

Um einen qualitativen Eindruck eines funktionalen Zusammenhanges zu gewinnen ist es sehr praktisch, diesen graphisch darzustellen. Das folgende Beispiel zeigt, dass wir diese Aufgabe  mit Python und einem passenden Paket erfüllen können, und zwar ohne großen Aufwand und innerhalb **einer** Umgebung.

Wenn Sidue am Ende dieses Notebooks alle Beispiele gemeinsam im Blick hast, hast du eine erste Idee davon, wie groß die Bedeutung der letzten Aussage ist: Wir müssen keine Daten und Funktionalitäten aus unterschiedlichsten Lösungen mit hohem manuellen Aufwand zusammenführen, sondern können alles in einer einzigen Umgebung erreichen. Dies spart langfristig enorm Zeit, und ist den (einmaligen) Mehraufwand dieses Grundlagenkurses in Python mehr als wert.

Wie genau die verwendeten Python-Pakete **numpy** und **matplotlib** funktionieren, erklären wir in einer späteren Einheit dieses Kurses. Auch kannst du im Moment noch syntaktische Details ignorieren.



```python
import numpy as np
import matplotlib.pyplot as plt

# erzeugt äquidistante Auswertungspunkte im gegebenen Intervall zwischen -pi und pi
eval_points = np.linspace(-np.pi, np.pi, 300)

# wertet die Sinusfunktion in allen Auswertungspunkten gleichzeitig aus
values = np.sin(eval_points)

# plottet die generierten (x,y)-Paare (Paare aus Punkten auf der x-Achse und 
# zugehörigen Funktionsauswertungen  auf der y-Achse) als Polygonzug / Linienzug
plt.plot(eval_points,values)                   
      
# zeigt den Plot an
plt.show()
```
