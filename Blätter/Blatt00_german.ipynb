## 1. Mathematische Grundoperationen
`Python` ist eine hervorragende Sprache für die Mathematik. Sie kann als einfacher Taschenrechner oder als leistungsstarkes Werkzeug für komplexe Ingenieursaufgaben dienen.

```python
10 + 5  # Addition 10 + 5 = 15
```

```python
20 - 7  # Subtraction 20 - 7 = 13
```

```python
4 * 3   # Multiplication 4 * 3 = 12 
```

```python
15 / 2  # Division 15 / 2 = 7.5 (notice the point separating the decimals)
```

Beachte, dass das Ergebnis eine Dezimalzahl ist und der Datentyp nun ein `float` ist. In `Python` ist ein `float` (kurz für „floating-point number“ oder Gleitkommazahl) ein Datentyp, der zur Darstellung reeller Zahlen mit einem Dezimalpunkt verwendet wird.

Warum ist das wichtig?

In Python führt die Standard-Division (`\`) immer zu einem `float`, auch wenn kein Rest bleibt. Das stellt sicher, dass Berechnungen so präzise wie möglich sind.

- **Integer**: Ganze Zahlen ohne Punkt (z. B. 5, -2, 1000).

- **Float**: Zahlen mit Dezimalpunkt (z. B. 5.0, -2.5, 3.14159).

```python
type(15/2) # Datentyp von 15/2
```

```python
10/2 # Die Operation '/' gibt immer Gleitkommazahlen zurück
```

```python
type(10/2) # Datentyp von 10/2
```

## 3. Weitere mathematische Operationen: Potenzen, Ganzzahl-Division und Modulo
Über die Grundrechenarten hinaus bietet Python spezielle Operatoren für Potenzen und besondere Arten der Division.

- **Potenz (`**`)**: Berechnet die Potenz einer Zahl ($2^3$)

- **Ganzzahl-Division (`//`)**: Dividiert und rundet das Ergebnis auf die nächste ganze Zahl ab (Floor Division).

- **Modulo (`%`)**: Gibt dir ausschließlich den Rest einer Division an.

```python
print(2 ** 3)  # 2 hoch 3 (2*2*2)
print(17 // 5) # Ganzzahl-Division: Wie oft passt die 5 in die 17?
print(17 % 5)  # Modulo: Was bleibt übrig, wenn man 17 durch 5 teilt?
```

## 4. Reihenfolge verschiedener Operationen
`Python` folgt den standardmäßigen mathematischen Vorrangregeln. Verwende Klammern, um sicherzustellen, dass bestimmte Teile deiner Prozedur zuerst ausgeführt werden.

```python
# Punkt vor Strich
print(49 * 2 + 7)   # Ergebnis 105

# Klammern zuerst
print((49 * 2) + 7) # Ergebnis 105
```

```python
# Addition vor Multiplikation
print(49 * (2 + 7))   # Ergebnis 441
```

**Ziel**: Wandle eine Temperatur mithilfe von Python-Variablen und Operatoren von Celsius in Fahrenheit um.

Die Formel:

$$F=(C \cdot 9/5)+32$$

**Anweisungen:**

1. **Daten:** Erstelle eine Variable c und weise ihr einen Wert zu (z. B. 25).

2. **Prozedur:** Erstelle eine Variable f und berechne die Umrechnung mithilfe der oben genannten Formel.

3. **Ausgabe:** Benutze den `print()`-Befehl, um das Endergebnis anzuzeigen.

```python
# Lösung
c = 25 
f = (c * 9 / 5) + 32
print(f)

# Answer: 77
```
