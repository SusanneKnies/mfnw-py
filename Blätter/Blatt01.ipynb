### Indizierung von Listen:
Die Elemente einer Liste sind in Python von 0 beginnend indiziert, bspw. wird die Liste <code>s = [1, 44, ['s', 'p'], '4', 5]</code> intern folgendermaßen indiziert:
<table style="font-size: 1em"><tr><th>Index</th><td>0</td><td>1</td><td>2</td><td>3</td><td>4</td></tr>
<th>-Index</th><td>-5</td><td>-4</td><td>-3</td><td>-2</td><td>-1</td></tr>
<tr><th>Wert</th><td>1</td><td>44</td><td>['s', 'p']</td><td>'4'</td><td>5</td></tr></table>

Damit kann direkt auf die Elemente der Liste zugegriffen werden, die negativen Indizes stellen dabei einen gewissen zyklischen Zugriff sicher, allerdings "nur einmal" und nicht komplett zyklisch:

```python
s = [1, 44, ['s', 'p'], '4', 5]
print("s[0] =", s[0])
print("s[2] =", s[2])
print("s[-4] =", s[-4])
print("s[-6] =", s[-6])  # Index 6 existiert nicht
```

```python
# Der Zugriff kann auch verschachtelt stattfinden, um auf Listen von Listen zuzugreifen:
s = [1, 44, ['s', 'p'], '4', 5]
print("s[2][1] =", s[2][1])
```

Gängige Listenoperationen können damit wie folgt selektiv ausgeführt werden:

<font color='gray'>Common list operations can thus be performed selectively as follows:</font>

```python
s = [1,2,3]
s[0] = -1   # Ändert den Wert an der esten Stelle zu -1
print(s)
s.append(4) # Hängt die 4 an das Ende der Liste an
print(s)
s.insert(4,1) # Fügt die Zahl 1 an der 4ten (eigentlich 5ten, wegen der Null-Indizierung) Stelle ein
print(s)
del s[2]    # Löscht das dritte (Index 2) Element der Liste
print(s)
del s[0:3]  # Es können auch mehrere Elemente auf einmal gelöscht werden
print(s)
```

### Extraktion von Teillisten
Python bietet mit dem Doppelpunkt-Operator eingebaute Mechanismen, um Teillisten zu extrahieren. Dabei gilt die folgende Regel: Bei allen Indexoperationen wird der Startindex stets dazugezählt, der letzte Index wird aber nicht beachtet. Das ist eine Konsequenz der Indizierung mit 0.

<font color='gray'> Extracting Sublists
Python provides built-in mechanisms for extracting sublists using the colon operator. The following rule applies: In all indexing operations, the starting index is always included, but the last index is ignored. This is a consequence of zero-based indexing. </font>

```python
s = [1, 44, ['s', 'p'], '4', 5]
print("s[0:2] =", s[0:2])
print("s[1,3] =", s[1:3])
print("s[0:len(s)] =", s[0:len(s)])
```

```python
s = [1, 44, ['s', 'p'], '4', 5]

# Wird das zweite Argument weggelassen, wird stets bis zum Ende iteriert:
print("s[0:] =", s[0:])  # Index 0,1,2,3,4
# Wird das erste Argument weggelassen, wird stets am Anfang begonnen
print("s[:3] =", s[:3])  # Index 0,1,2
```

```python
s = [1, 44, ['s', 'p'], '4', 5]

# Mit dem dritten Argument wird festgelegt, in welcher Schrittweite iteriert wird:
# Wir extrahieren in Zweierschritten aus der Liste:
print("s[0:len(s):2] =", s[0:len(s):2])
# Und die dazu passende abkürzende Schreibweise:
print("s[::2] =", s[::2])
```

Die Elemente der Liste können auch von hinten beginnend indiziert werden:

<font color='gray'>The elements in the list can also be indexed starting from the end:</font>

```python
s = [1, 44, ['s', 'p'], '4', 5]
# Letztes Element:
print(s[-1])
# Letzten 3 Elemente:
print(s[-3:])
# Zweites bis vorletztes Element:
print(s[1:-1])
```

In diesem Beispiel simulieren wir die Vererbung von Genvarianten (**Allelen**). Wir nehmen an, wir haben zwei Elternteile mit bestimmten Genotypen und möchten alle möglichen Kombinationen für deren Nachkommen berechnen.

Ein zentrales Problem in der Programmierung ist hierbei die **Redundanz**: Wenn ein Kind die Allele `'A'` und `'C'` erbt, ist das Ergebnis (`'AC'`) identisch mit der Kombination `'C'` und `'A'`. In der Informatik nutzen wir daher Techniken, um solche Duplikate zu bereinigen und nur die einzigartigen Möglichkeiten anzuzeigen.

**Der Prozess im Überblick:**

- **1. Kombinatorik:** Wir paaren jedes Allel von Elternteil 1 mit jedem Allel von Elternteil 2.

- **2. Sortierung:** Wir sortieren die Buchstaben innerhalb der Paare (aus CA wird AC), damit sie vergleichbar werden.

- **3. Filterung:** Wir entfernen doppelte Einträge (z. B. mit einem set), um die tatsächliche Vielfalt der Nachkommen zu sehen.

In this example, we simulate the inheritance of gene variants (**alleles**). We assume we have two parents with specific genotypes and want to calculate all possible combinations for their offspring.

A key issue in programming here is **redundancy**: If a child inherits the alleles `'A'` and `'C'`, the result (`'AC'`) is identical to the combination `'C'` and `'A'`. In computer science, we therefore use techniques to eliminate such duplicates and display only the unique possibilities.

**An overview of the process:**

- **1. Combinatorics:** We pair each allele from Parent 1 with each allele from Parent 2.

- **2. Sorting:** We sort the letters within the pairs (CA becomes AC) so that they can be compared.

- **3. Filtering:** We remove duplicate entries (e.g., using a set) to see the actual diversity of the offspring.
</font>

### 1. Kombinatorik

Zuerst definieren wir zwei Strings, `'CT'` und `'AC'`, die jeweils die Allele der beiden Elternteile repräsentieren. Um alle möglichen Genotypen der Nachkommen zu ermitteln, nutzen wir eine **List-Comprehension**.

Dabei verschachteln wir zwei for-Schleifen, die nacheinander durch beide Strings iterieren. Das Programm kombiniert so jedes Allel des ersten Elternteils mit jedem Allel des zweiten. Das Ergebnis ist eine Liste, die alle theoretisch denkbaren Kombinationen für die nächste Generation enthält.

<font color='gray'>

### 1. Combinatorics

First, we define two strings, `'CT'` and `'AC'`, which represent the alleles of the two parents, respectively. To determine all possible genotypes of the offspring, we use a **list comprehension**.

To do this, we nest two for loops that iterate sequentially through both strings. The program thus combines each allele of the first parent with each allele of the second. The result is a list containing all theoretically possible combinations for the next generation.
</font>

```python
parent1 = 'CT'
parent2 = 'AC'

möglichkeiten = [a+b for a in parent1 for b in parent2]
print(möglichkeiten)
```

### 2. Allele vereinheitlichen (Sortierung)

Da wir die Allele beider Elternteile nacheinander zusammengefügt haben, enthält unsere Liste aktuell Kombinationen in der Reihenfolge ihrer Entstehung – zum Beispiel `'CA'` (C von Elternteil 1 und A von Elternteil 2).

In der Genetik spielt die Reihenfolge innerhalb eines Genotyps jedoch meist keine Rolle. Um die Ergebnisse vergleichbar zu machen und später Duplikate leichter erkennen zu können, bringen wir die Buchstaben in eine **alphabetische Reihenfolge** (also AC statt CA).

**Der Prozess:**
- Wir betrachten jedes Kind in unserer Liste einzeln.
- Wir sortieren die Buchstaben innerhalb des Strings alphabetisch.
- Das Ergebnis speichern wir in einer neuen Liste ab.

<font color='gray'>

### 2. Standardize Alleles (Sorting)

Since we have combined the alleles from both parents one after the other, our list currently contains combinations in the order in which they were formed—for example, `'CA'` (C from Parent 1 and A from Parent 2).

In genetics, however, the order within a genotype usually does not matter. To make the results comparable and to make it easier to identify duplicates later, we arrange the letters in **alphabetical order** (i.e., AC instead of CA).

**The Process:**
- We examine each child in our list individually.
- We sort the letters within the string alphabetically.
- We store the result in a new list.
</font>

```python
# Code für eine einzelne Allele

b = 'CA'

# Der Befehl sorted() macht aus einem string eine Liste. 
# Diese Liste enthält dann alle Zeichen (characters) des strings in alphabetischer Reihenfolge
b_sorted = sorted(b)
print(b_sorted)

# Der Befehl join() fügt eine Liste wieder zu einem string zusammen.
# Der string vor dem join Befehl legt fest, welcher string zwischen die einzelnen Listen-Elemente gesetzt werden soll.
b_sorted = ''.join(b_sorted)
print(b_sorted)
```

```python
# join und sorted Befehl hintereinander ausführen
# List-comprehension: 'for' sorgt dafuer, dass alle Listen-Elemente nacheinander durchgegangen werden.
möglichkeiten = [''.join(sorted(allel)) for allel in möglichkeiten]
print(möglichkeiten)
```

### 3. Doppelte Möglichkeiten aussortieren
In unserem bisherigen Beispiel mit den Eltern `'AC'` und `'CT'` waren alle vier Kombinationen bereits unterschiedlich. Doch was passiert, wenn beide Elternteile den gleichen Genotyp haben, zum Beispiel beide `'CT'`?

In diesem Fall würden durch die Paarung Kombinationen mehrfach entstehen (z. B. CT aus C von Mutter / T von Vater und noch einmal CT aus T von Mutter / C von Vater nach der Sortierung). Um eine saubere Übersicht der tatsächlichen genetischen Vielfalt zu erhalten, müssen wir diese Redundanzen entfernen.

Die Lösung: Das `set`
In Python ist ein Set (Menge) ein Datentyp, der jedes Element nur genau einmal speichert. Verwandeln wir unsere Liste in ein Set, werden alle doppelten Ergebnisse automatisch gelöscht.

<font color='gray'>

### 3. Eliminating Duplicate Possibilities
In our previous example with the parents `'AC'` and `'CT'`, all four combinations were already distinct. But what happens if both parents have the same genotype, for example, both `'CT'`?

In this case, the mating would result in combinations appearing multiple times (e.g., CT from C from mother / T from father and, after sorting, CT again from T from mother / C from father). To get a clear picture of the actual genetic diversity, we need to remove these duplicates.

The solution: The `set`
In Python, a set is a data type that stores each element exactly once. If we convert our list into a set, all duplicate entries are automatically removed.
</font>

```python
parent3 = 'CT'

# Möglichkeiten von CT und CT erzeugen und alphabetisch sortiert in Liste speichern
möglichkeiten2 = [''.join(sorted(a+b)) for a in parent1 for b in parent3]
print(möglichkeiten2)

# Liste in set() umwandeln. Eine Menge erkennt man an den geschweiften Klammern {}
möglichkeiten2 = set(möglichkeiten2)
print(möglichkeiten2)

# set wieder in Liste umwandeln
möglichkeiten2 = list(möglichkeiten2)
print(möglichkeiten2)
```
