 ## 1. Vektoren als Tupel
 In Python lassen sich Vektoren als Tupel darstellen. Ein Tupel ist eine unveränderliche Sequenz von Elementen, die in runden Klammern () geschrieben wird.

```python
# Erstellung von Vektoren als Tupel
v1 = (1, 3, 2)
v2 = (5, 3, 4)
```

Um die **Differenz** zweier Vektoren zu berechnen, subtrahiert man die entsprechenden Elemente. Tupel unterstützen diese Operation nicht direkt. Dafür ist die ``zip()``-Funktion hilfreich. Sie kombiniert Elemente aus zwei oder mehr iterierbaren Objekten paarweise. ``zip(v1, v2)`` erzeugt Paare der Elemente: (1, 5), (3, 3) und (2, 4).

Anschließend wird eine Generator-Expression ``(a-b for a,b in zip(v1,v2))`` zusammen mit ``tuple()`` verwendet. Diese effiziente Ausdrucksweise erzeugt eine neue Sequenz, indem für jedes Paar (a, b) die Subtraktion a-b durchgeführt wird.

```python
# Differenz der beiden Vektoren
diff_v = tuple(a - b for a, b in zip(v1, v2))
print(diff_v)
```

## 2. Vektoren mit Numpy
Die Subtraktion von Vektoren ist in **NumPy**, einer der wichtigsten Bibliotheken für wissenschaftliches Rechnen in Python, besonders einfach. Im Gegensatz zu den Standard-Python-Tupeln, bei denen eine elementweise Subtraktion manuell über eine Schleife oder eine ``zip``-Funktion erfolgen muss, erledigt NumPy dies automatisch und effizient.

TBD: Numpy muss auch installiert werden

```python
import numpy as np

# Vektoren als numpy-array erstellen
u1 = np.array([1, 3, 2])
u2 = np.array([5, 3, 4])

diff_u = u1-u2
print(diff_u)
```

## 3. Vorteile von Numpy
Ein wesentlicher Vorteil von NumPy ist seine **Fehlererkennung**. Während Python-Tupel stillschweigend Fehler bei der Dimensionierung übersehen können, indem sie einfach die kürzeren Vektoren ignorieren, schlägt NumPy sofort Alarm.

Das folgende Beispiel demonstriert dies: Vektor v1 hat 3 Elemente, Vektor v3 hat 4. Bei Verwendung von zip() wird das vierte Element von v3 einfach ignoriert, da es keinen passenden Partner in v1 findet. Das Ergebnis ist rechnerisch fehlerhaft.

```python
v3 = (5, 2, 6, 2)
print("Differenz mit Tupeln:", tuple(a-b for a,b in zip(v1,v3)))
```

NumPy erkennt derartige Dimensionsunterschiede sofort und löst einen ValueError aus. Dieser Mechanismus hilft dabei, Fehler im Code frühzeitig zu erkennen:

```python
u3 = np.array([5,2,6,2])
print("Differenz mit Numpy:", u1-u3)  # → Fehler!
```

## 5. Vektoren zeichnen
Mit **Matplotlib** kannst du Vektoren im Koordinatensystem darstellen.

```python
import matplotlib.pyplot as plt

origin = (0, 0)

x1 = np.array([1, 2])
x2 = np.array([0, -1])
diff_x = x2 - x1

plt.quiver(*origin, *x1, angles='xy', scale_units='xy', scale=1, color='r', label="x1")
plt.quiver(*origin, *x2, angles='xy', scale_units='xy', scale=1, color='b', label="x2")
plt.quiver(*origin, *diff_x, angles='xy', scale_units='xy', scale=1, color='g', label="x2-x1")

plt.xlim(-6, 6)
plt.ylim(-6, 6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()

plt.show()
```

### Menge plotten in python

```python
x = np.linspace(-1.5, 1.5, 300)
y = np.linspace(-1.5, 1.5, 300)

points = [(a, b) for a in x for b in y]
distances = [np.sqrt(a**2 + b**2) for a, b in points]
colors = ['red' if d <= 1 else 'blue' for d in distances]

xs, ys = zip(*points)
plt.scatter(xs, ys, c=colors, s=5)
# plt.gca().set_aspect("equal")  # Kreise sollen rund sein
plt.show()
```
