# Python math basics
In computing, we can define **Data** as the objects we want to manipulate, and **Procedures** as the descriptions of the rules for manipulating those data. The simpler datum we can consider is a given number.

```python
365 # This is a number (and the rest of the line is a comment)
```

In Python, `int` stands for **integer**. An integer is a whole number, positive or negative, without decimals, of unlimited length. It is one of the most fundamental data types in this language.

## 1. The `print()` command
To see the result of a calculation in `Python`, we use the `print()` command. 
Think of it as a way to tell the computer: "Show me the result on the screen!"


```python
print(10) # Not so different from previous codelines.
```

```python
# But the point is that EVERYTHING inside the parentheses will be displayed!
print(100) 
print(10 + 5)
print("Hello! The result of 2 + 2 is:")
print(2 + 2)
```

Notice that the text inside the `" "` is called a **String** (often abbreviated as `str`). It is simply a sequence of characters (letters, numbers, symbols, or spaces) that the computer treats as literal text rather than as a command or a number.

```python
type("This is a string of text. It may contains numbers like 1, -46 or 102393.")
```

## 2. More math operations: Exponents, Floor division, and Modulo
Beyond basic math, Python has special operators for powers and specific types of division.

* **Exponent (`**`)**: Power of a number ($2^3$).
* **Floor Division (`//`)**: Divides and rounds down to the nearest whole number.
* **Modulo (`%`)**: Gives you only the **remainder** of a division.

```python
print(2 ** 3)  # 2 to the power of 3 (2*2*2)
print(17 // 5) # Floor division: how many times 5 fits into 17
print(17 % 5)  # Modulo: the remainder left over after dividing 17 by 5
```

## 3. Order of operations
Python follows standard mathematical precedence rules: use parentheses to ensure specific parts of your procedure are executed first.

```python
# Multiplication first
print(49 * 2 + 7)   # Results in 105

# Parentheses first
print((49 * 2) + 7) # Also 105, but clearer
```

```python
# Addition first
print(49 * (2 + 7))   # Results in now 441
```

## 4. Variables: Storing Data
A variable is a named location used to store data in the computer's memory. You assign a value to a variable using the = operator. This allows you to reuse data without re-typing it.

```python
a = 100
print(a)
```
