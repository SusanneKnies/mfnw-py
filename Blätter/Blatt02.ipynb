- `.shape`: Die Dimensionen des Arrays (gibt bei einem Vektor die Anzahl der Elemente an).

<font color='gray'>

### Important Properties of a Vector

Every NumPy vector has certain attributes that provide information about its structure:

- `.shape`: The dimensions of the array (for a vector, this indicates the number of elements).
- `.size`: The total number of elements in the array.
</font>

```python
print(u1.shape)
print(u1.size)
```

### Addition und Subtraktion

Bei der Addition oder Subtraktion von zwei Vektoren werden die Elemente an den jeweils gleichen Positionen miteinander verrechnet.

**Voraussetzung:** Beide Vektoren müssen die gleiche Länge (Shape) haben.

<font color='gray'>

### Addition and Subtraction

When adding or subtracting two vectors, the corresponding components are added or subtracted.

**Prerequisite:** Both vectors must have the same magnitude (shape).
</font>

```python
u2 = np.array([6,2,3])

print(u1-u2)
print(u1+u2)
```

### Multiplikation mit einem Skalar

Ein Skalar ist in diesem Kontext einfach eine einzelne Zahl (z. B. ein int oder float). Wenn du einen Vektor mit einem Skalar multiplizierst, wird jedes einzelne Element des Vektors mit dieser Zahl multipliziert. In der Geometrie entspricht dies einer Streckung oder Stauchung des Vektors.

<font color='gray'>

### Multiplication by a Scalar

In this context, a scalar is simply a single number (e.g., an int or float). When you multiply a vector by a scalar, each individual element of the vector is multiplied by that number. In geometry, this corresponds to stretching or compressing the vector.
</font>

```python
print(2*u1)
print(u2/4)
```

### Multiplikation von zwei Vektoren: Elementweise Multiplikation

In NumPy bewirkt der Stern-Operator `*`, dass die Elemente paarweise an den jeweiligen Positionen multipliziert werden. Das Ergebnis ist wieder ein Vektor. Mathematisch nennt man dies das Hadamard-Produkt.

<font color='gray'>

### Multiplication of two Vectors: Element-wise Multiplication

In NumPy, the asterisk operator `*` multiplies the elements pairwise at their respective positions. The result is again a vector. Mathematically, this is called the Hadamard product.
</font>


Beispiel:
$$
\begin{pmatrix} 
1 \\
2
\end{pmatrix}​	
\cdot
\begin{pmatrix}
3\\
4
\end{pmatrix}
=
\begin{pmatrix}
1⋅3\\
2⋅4
\end{pmatrix}
=
\begin{pmatrix}
3\\
8
\end{pmatrix}
$$

```python
print(np.array([1,2])*np.array([3,4]))
```

### Multiplikation von zwei Vektoren: Das Skalarprodukt
Das Skalarprodukt multipliziert ebenfalls die Elemente paarweise, addiert diese Produkte aber im Anschluss auf. Das Ergebnis ist eine einzelne Zahl (ein Skalar). In der Geometrie nutzt man dies, um Winkel zwischen Vektoren zu berechnen oder Projektionen zu bestimmen.

<font color='gray'>

### Multiplication of Two Vectors: The Dot Product
The dot product also multiplies the elements in pairs, but then adds these products together. The result is a single number (a scalar). In geometry, this is used to calculate angles between vectors or to determine projections.
</font>


Beispiel:
$$
\begin{pmatrix} 
1 \\
2
\end{pmatrix}​	
\cdot
\begin{pmatrix}
3\\
4
\end{pmatrix}
=(1 \cdot 3)+(2\cdot 4)=3+8=11
$$

```python
print(np.dot(np.array([1,2]), np.array([3,4])))
```

### Aufgabe 2
Berechne den **Betrag (die Länge)** eines Vektors.
Formel:  

$$\|v\| = \sqrt{v_1^2 + v_2^2 + \dots + v_n^2}$$

```python
# Tipp: np.linalg.norm() macht das schon für dich.
# Versuche aber zuerst, die Summe der Quadrate selbst zu berechnen.

# Lösung
v = np.array([2,-1,2])

norm1=np.sqrt(np.dot(v,v))
norm2=np.sqrt(sum(v*v))
norm3=np.linalg.norm(v)

print(norm1, norm2, norm3)
```

### Example for Advanced Users

- **Note:** This section is an optional in-depth exploration for anyone who already has some experience or is simply curious. The content is not strictly necessary for the rest of the course—it’s designed to expand your knowledge if you’re up for a little challenge!

With **Matplotlib**, you can plot vectors in a coordinate system.
</font>

```python
import matplotlib.pyplot as plt

# Definieren des Ursprungs (Startpunkt der Vektoren bei 0,0)
origin = (0, 0)

# Erstellen von zwei Beispiel-Vektoren als NumPy-Arrays
x1 = np.array([1, 2])
x2 = np.array([0, -1])

# Berechnung des Differenzvektors (Vektor von der Spitze von x1 zur Spitze von x2)
diff_x = x2 - x1

# Zeichnen der Vektoren mit der Quiver-Funktion (Pfeil-Darstellung)
# *origin entpackt die Koordinaten (0,0) als Startpunkt
# angles='xy' und scale_units='xy' sorgen dafür, dass die Pfeile korrekt zum Koordinatensystem passen
plt.quiver(*origin, *x1, angles='xy', scale_units='xy', scale=1, color='r', label="x1")
plt.quiver(*origin, *x2, angles='xy', scale_units='xy', scale=1, color='b', label="x2")

# Zeichnen des Differenzvektors (hier vom Ursprung aus visualisiert)
plt.quiver(*origin, *diff_x, angles='xy', scale_units='xy', scale=1, color='g', label="x2-x1")

# Festlegen der Achsenabschnitte (von -6 bis 6 auf beiden Achsen)
plt.xlim(-6, 6)
plt.ylim(-6, 6)

# Zeichnen der x- und y-Achsenlinien in Schwarz
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Sicherstellen, dass die Achsen den gleichen Maßstab haben (1 Einheit x = 1 Einheit y)
# Ohne dies würden Kreise wie Ellipsen aussehen und Vektor-Winkel verzerrt werden
plt.gca().set_aspect('equal', adjustable='box')

# Anzeigen der Legende basierend auf den 'label'-Parametern
plt.legend()

# Plot generieren und anzeigen
plt.show()
```
