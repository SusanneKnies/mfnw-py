Auf Blatt 1 hattest du bereits gelernt, wie man Vektoren mit numpy erstellen kann. Ebenso kannst du auch Matrizen als numpy Objekt erstellen. Zuerst importieren wir hierfür wieder numpy:

```python
import numpy as np
```

Nun erstellen wir den Vektor $b$ und die Matrix $A$. Wie man einen `numpy`-Vektor erstellt weißt du bereits. Um eine `numpy`-Matrix zu erstellen verwenden wir eine Liste, die mehrere Listen enthält. Diese Listen repräsentieren dabei die *Zeilen* der Matrix. 

```python
b = np.array([3, 4, 1])
A = np.array([[6, -1, 3], [7, -2, -2], [-5, 2, 4]])

print('b:', b)
print('A:', A)
```

Damit A auch als Matrix eingelesen werden kann, muss jede Liste gleich viele Zahlen beinhalten. Ist dies nicht der Fall, erhalten wir folgende Fehlermeldung:

```python
A = np.array([[3,4], [2,5,3], [8,3,5]])
```

### Matrix-Vektor Multiplikation

Im folgenden möchten wir das Produkt der Matrix $A$ mit dem Vektor $b$ berechnen. Intuitiv könnte man hier `A*b` ausrechnen. Dieser Befehl liefert auch ein Ergebnis, jedoch entspricht das nicht dem Matrix-Vektor-Produkt! Wenn du in Python mit NumPy arbeitest, führt der Ausdruck A * b eine elementweise Multiplikation durch (auch bekannt als Element-wise Multiplication oder Hadamard-Produkt).

Das Ergebnis ist eine Matrix derselben Form wie $A$, bei der jedes Element $a_{ij}$ mit dem entsprechenden Element $b_j$ multipliziert wurde.

```python
A*b
```

Um die Matrix-Vektor-Multiplikation richtig durchzuführen haben wir verschiedene Möglichkeiten:

```python
c = A @ b
d = np.matmul(A, b)
e = np.dot(A,b)

print(c, d, e)
```

### Lineare Gleichungssysteme

Wir können mit `numpy` aus LGS der Form $A \cdot x = b$ lösen. Hierfür wird der Befehl `np.linalg.solve()` verwendet.

```python
x = np.linalg.solve(A, b)

print('x:', x)

# Zur Überprüfung
print('A@x:', A@x)
print('b:', b)
```

`np.linalg.solve()` kann nur für eindeutig lösbare LGS verwendet werden:

```python
B = np.array([[-6, 9, -12], [2, -3, 4], [4, -6, 8]])
c = np.array([-20, 7, 12])

np.linalg.solve(B,c)
```
