## 1. Datentypen umwandeln (Type Casting)
In `Python` ist es entscheidend, welchen Datentyp eine Variable besitzt. Verschiedene Datentypen verhalten sich unterschiedlich und können nicht immer direkt miteinander kombiniert werden. Deshalb ist es oft notwendig, Datentypen bewusst umzuwandeln, bevor man mit ihnen arbeitet. `Python` ist stark typisiert – das bedeutet, dass der Interpreter genau darauf achtet, welche Typen verwendet werden, und bestimmte Operationen zwischen unpassenden Datentypen nicht erlaubt sind.

**Das Problem: Der Typ-Konflikt**

Versuchst du, einen Text und eine Zahl direkt zu kombinieren, bricht `Python` mit einem TypeError ab:

```python
alter = 20
text = "Ich bin " + alter + " Jahre alt." # Fehler!
```

Der Computer weiß nicht, ob er die Zahl in Text verwandeln oder den Text mathematisch addieren soll.

**Die Lösung: Explizite Umwandlung**

Wir müssen Python genau sagen, welcher Typ gewünscht ist. Hierfür gibt es drei wichtige Funktionen:

1. `str(wert)`: Wandelt alles in einen String um (ideal für Ausgaben).

2. `int(wert)`: Wandelt in eine Ganzzahl um (schneidet Nachkommastellen bei Floats einfach ab!).

3. `float(wert)`: Wandelt in eine Kommazahl um.

```python
# Richtiges Kombinieren durch Umwandlung
alter = str(alter)
ausgabe = "Ich bin " + alter + " Jahre alt."
print(ausgabe) # Funktioniert!
print(type(alter))
```

Die Variable `alter` wurde in einen String umgewandelt. Dadurch kann sie nicht mehr für mathematische Berechnungen verwendet werden, sondern nur noch als Text.

```python
print(alter+5)
```

```python
# alter wieder in int verwandeln (float wäre auch möglich)
alter = int(alter)
print(type(alter))
print(alter+5)
```

Allerdings lassen sich nicht alle Datentypen beliebig ineinander umwandeln. Ein String kann fast immer erzeugt werden, da sich nahezu jeder Wert in eine Textdarstellung umwandeln lässt. Umgekehrt ist das nicht immer möglich: Der Versuch, einen nicht-numerischen String wie `int('a')` in eine Zahl umzuwandeln, führt zu einer Fehlermeldung, da der Inhalt nicht als Zahl interpretiert werden kann.

```python
int('hallo')
```
