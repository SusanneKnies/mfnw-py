## 1. Datentypen umwandeln (Type Casting)
In `Python` ist es entscheidend, welchen Datentyp eine Variable besitzt. Verschiedene Datentypen verhalten sich unterschiedlich und können nicht immer direkt miteinander kombiniert werden. Deshalb ist es oft notwendig, Datentypen bewusst umzuwandeln, bevor man mit ihnen arbeitet. `Python` ist stark typisiert – das bedeutet, dass der Interpreter genau darauf achtet, welche Typen verwendet werden, und bestimmte Operationen zwischen unpassenden Datentypen nicht erlaubt sind.

**Das Problem: Der Typ-Konflikt**

Versuchst du, einen Text und eine Zahl direkt zu kombinieren, bricht `Python` mit einem TypeError ab:

```python
alter = 20
text = "Ich bin " + alter + " Jahre alt." # Fehler!
```
