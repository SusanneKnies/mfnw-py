## 3. Weitere mathematische Operationen: Potenzen, Ganzzahl-Division und Modulo
Über die Grundrechenarten hinaus bietet Python spezielle Operatoren für Potenzen und besondere Arten der Division.

- **Potenz (`**`)**: Berechnet die Potenz einer Zahl ($2^3$)

- **Ganzzahl-Division (`//`)**: Dividiert und rundet das Ergebnis auf die nächste ganze Zahl ab (Floor Division).

- **Modulo (`%`)**: Gibt dir ausschließlich den Rest einer Division an.

<font color='gray'> 3. More math operations: Exponents, Floor division, and Modulo
Beyond basic math, Python has special operators for powers and specific types of division.

* **Exponent (`**`)**: Power of a number ($2^3$).
* **Floor Division (`//`)**: Divides and rounds down to the nearest whole number.
* **Modulo (`%`)**: Gives you only the **remainder** of a division. </font>

```python
print(2 ** 3)  # 2 to the power of 3 (2*2*2)
print(17 // 5) # Floor division: how many times 5 fits into 17
print(17 % 5)  # Modulo: the remainder left over after dividing 17 by 5
```

## 4. Reihenfolge verschiedener Operationen
`Python` folgt den standardmäßigen mathematischen Vorrangregeln. Verwende Klammern, um sicherzustellen, dass bestimmte Teile deiner Prozedur zuerst ausgeführt werden.

<font color='gray'> 4. Order of operations
Python follows standard mathematical precedence rules: use parentheses to ensure specific parts of your procedure are executed first.</font>

```python
# Multiplication first
print(49 * 2 + 7)   # Results in 105

# Parentheses first
print((49 * 2) + 7) # Also 105, but clearer
```

```python
# Addition first
print(49 * (2 + 7))   # Results in now 441
```

Zuerst einmal laden wir dafür die Bibliotheken `numpy` und `matplotlib`. Diese stellen Funktionen zur Verfügung, welche nicht standardmäßig in `python` implementiert sind. Um sie zu laden verwendest du folgende Befehle:

<font color='gray'>First, we’ll import the `numpy` and `matplotlib` libraries. These libraries provide functions that aren’t included in Python by default. To import them, use the following commands:</font>

```python
import matplotlib.pyplot as plt
import numpy as np
```

Achte darauf den ersten Code-Block auch auszuführen, um die Bibliotheken zu laden. Mit `as np` legen wir fest, dass die in numpy verfügbaren Funktionen mit `np.` abgerufen werden können. Nun können wir folgenden Code-Block ausführen:

<font color='gray'>Be sure to run the first code block as well to load the libraries. With `as np`, we specify that the functions available in NumPy can be accessed using `np.`. Now we can run the following code block:</font>

```python
x = np.linspace(-1, 1, 100) # Divide the x-axis into 100 points between -1 and 1
```

Die in `numpy` verfügbare Funktion `linspace` erstellt gibt 100 geordnete Punkte zwischen -1 und 1 zurück, welche den gleichen Abstand zueiander haben.  Um nun die Funktion $f(x)=x^2$ kannst du folgenden Code verwenden:

<font color='gray'>The `linspace` function available in `numpy` returns 100 ordered points between -1 and 1 that are equally spaced.  To generate the function $f(x)=x^2$, you can use the following code:</font>

```python
y = x**2 # Calculate the y-value corresponding to each x-value

plt.plot(x,y, color='blue', label='f(x)=x^2') # plot the points
plt.xlabel("x") # label x-axis
plt.ylabel("y")
plt.legend()
plt.show() # The plot below is displayed only after this command is executed
```

Hier findest du denselben Code. Versuche nun den Code so zu ändern, dass er folgende Bedingungen erfüllt:

- Der Graph zeigt die Funktion $f(x)=2x^3-x$.

<font color="gray">Here is the same code. Now try modifying the code so that it meets the following conditions:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 100)
y = x**2

plt.plot(x,y, color='blue', label='f(x)=x^2')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
